## Polynomial Linear regression

- When relationship between dependent and independent variables is non linear then we build polynomial regression. but why it is called polynomial linear regression?

It is linear because the coefficients are still linear.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

dataset = pd.read_csv('./Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

## Linear regression to compare

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
reg
```

## Polynomial regression 

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
lin_reg2
```

#### Visualizing the linear regression model results

```python
plt.scatter(X, y, color='red')
plt.plot(X, reg.predict(X), color='blue')
plt.title("Truth or Bluff(Linear regression)")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(X_poly), color='blue')
plt.title("Truth or Bluff(Linear regression)")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
## Predicting salary information using linear regression
print(reg.predict([[6.5]]))
```

```python
## Predicting salary information using polynomial regression
print(lin_reg2.predict(poly.fit_transform([[6.5]])))
```
